```
#the image to be detected is first loaded into im
im = cv2.imread('image.jpeg')
```

The above figure is 8x8 grid. 
The task is to detect tiles with faces

```
#converting rgb image to gray image for easy detection and analysis 
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
```

```
cv2_imshow(gray)
```

```
cascPath = "haarcascade_frontalface_default.xml"
```

```
# Create the haar cascade
#Haar Cascade is deep learning algorithm that implements face recognition
faceCascade = cv2.CascadeClassifier(cascPath)
```

```
#detecting faces in the image 
faces = faceCascade.detectMultiScale(gray, 
                                     scaleFactor = 1.1,
                                     minNeighbors = 5,
                                     minSize=(30,30), 
                                     flags = cv2.CASCADE_SCALE_IMAGE)
```

```
#coordinates of all the face image detected(x,y,w,h)
faces
```

```
print("Found {0} faces".format(len(faces)))
```

Using cascade classifier, we could determine the number of face tiles in the image which is found to be 19!

```
#drawing rectangular boxes for faces
for (x, y, w, h) in faces:
     cv2.rectangle(im, (x, y), (x+w, y+h), (0, 255, 0), 2)

```

```
#gives a clear display of image in which the tiles with face are highlighted
cv2_imshow(im)
```
